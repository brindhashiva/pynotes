```python
#Chat Loop
while True:
    user = input("You: ")

    if user.lower() == "exit":
        print("Chat ended")
        break

    prompt = f"""You are a helpful AI assistant.
Answer the user's question clearly and shortly.

User: {user}
Assistant:"""

    response = llm(
        prompt,
        max_tokens=80,
        temperature=0.7,
        stop=["User:", "\n\n"]
    )

    text = response["choices"][0]["text"].strip()

    if text == "":
        print("Bot: No response generated")
    else:
        print("Bot:", text)
```
