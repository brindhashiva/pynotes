```python
nums = [1, 2, 3, 4, 5]
even = []
for n in nums:
    if n % 2 == 0:
        even.append(n)
print(even)
```

```python
text = "hello world"
count = 0
for ch in text:
    if ch == "l":
        count += 1
print("Count of l:", count)
```
